```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python

# Step 1: Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
```

```python

# Step 2: Explore the dataset
print(iris_df.head())
```

```python

# Step 3: Preprocess the data
X = iris_df.drop('target', axis=1)
```

```python

# Step 4: Perform K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
```

```python

# Step 5: Visualize the clusters
iris_df['cluster'] = labels

plt.figure(figsize=(12, 8))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('K-means Clustering of Iris Dataset')
plt.show()
```

```python

# Step 6: Evaluate the clustering
comparison = pd.crosstab(iris_df['target'], iris_df['cluster'], rownames=['Actual'], colnames=['Cluster'])
print(comparison)
```
